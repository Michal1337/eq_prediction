```python
import tensorflow as tf
import numpy as np
```

```python
#loads model
model = tf.keras.models.load_model("../models/test_model.keras")
model = tf.keras.models.load_model("../models/test_model.h5")
model = tf.keras.models.load_model("../models/test_model")
```

```python
# example input need to be in shape (None, block_size, n_features). In this example block_size = 32, n_features = 5 (mag, x, y, depth , diff_days). None is the batch dimension, which can be any size. As long as it fits in the VRAM.
input = np.random.rand(1, 32, 5)
```

```python
logits = model.predict(input)
logits
```

```python
# model return logits, after using softmax, we get the probabilities
probabilities = tf.nn.softmax(logits)
probabilities
```

```python
# first element is the probability of the first class, second element is the probability of the second class, here first class is no earthquake, second class is earthquake
```
